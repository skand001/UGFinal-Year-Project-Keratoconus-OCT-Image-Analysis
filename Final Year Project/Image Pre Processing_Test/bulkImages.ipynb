```python
import os
import re
import matplotlib.pyplot as plt
from PIL import Image
```

```python


data_path = '/mnt/scratch/mehphoenix/Images'

raw_images = []
labels = []

# loop through all the subdirectories in the main directory of any files, whose extensions end with two digits.
for subdir, dirs, files in os.walk(data_path):
    for file in files:
        if re.search(r"\.\d{2}$", file):
            raw_images.append(os.path.join(subdir, file))

print(f"The total number of raw images in the array 'raw_images' is: {len(raw_images)}")
```

```python
# show the first five images
for i in range(10):
    image_path = raw_images[i]
    image = Image.open(image_path)
    plt.imshow(image)
    # Get the resolution of the image
    width, height = image.size

    # Set the yticks to display pixels vertically by every 50 starting from the bottom
    plt.yticks(range(0, height + 1, 100))
    plt.show()
```
